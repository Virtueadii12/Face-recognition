```python
import cv2
```

```python
# Loading cascade file
face_cap = cv2.CascadeClassifier("/Users/adityasingh/haarcascade_frontalface_default.xml")

# Open webcam
video_cap = cv2.VideoCapture(0)

if not video_cap.isOpened():
    print("Error: Could not open webcam")
else:
    while True:
        ret, video_data = video_cap.read()
        if not ret:
            print("Failed to grab frame")
            break

        # Converting to grayscale for black & white 
        col = cv2.cvtColor(video_data, cv2.COLOR_BGR2GRAY)

        # Detecting faces
        faces = face_cap.detectMultiScale(
            col,
            scaleFactor=1.1,
            minNeighbors=5,
            minSize=(30, 30)
        )

        # Drawing Green rectangles to faces
        for (x, y, w, h) in faces:
            cv2.rectangle(video_data, (x, y), (x + w, y + h), (0, 255, 0), 2)

        cv2.imshow("Face Detection", video_data)

        # Press 'a' to exit
        if cv2.waitKey(1) & 0xFF == ord('a'):
            break

video_cap.release()
cv2.destroyAllWindows()
```
